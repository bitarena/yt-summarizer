# Summarizing with Basic Prompt

```python
from langchain.chat_models import ChatOpenAI
from langchain_community.document_loaders import YoutubeLoader
from langchain.schema import (
    AIMessage,
    HumanMessage,
    SystemMessage
)
from google.colab import userdata
import os

os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')

loader = YoutubeLoader.from_youtube_url(
    "https://www.youtube.com/watch?v=8Wgcbw_q6N0", add_video_info=True
)

result = loader.load()[0]
video_length = result.metadata['length']
transcript = result.page_content

messages = [
    SystemMessage(content='You are an expert copywriter with expertize in summarizing transcripts'),
    HumanMessage(content=f'Please provide a short and concise summary of the following transcript:\n: {transcript}')
]

llm = ChatOpenAI(
    temperature=0,
    model_name='gpt-3.5-turbo')

llm.get_num_tokens(transcript)
summary_output = llm.invoke(messages)
print(summary_output.content)
```
